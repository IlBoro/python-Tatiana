# Operating Systems

Программы используемые в работе  взаимодействуют с разными частями железа в ПК. Например с сетевой картой(INTEL, D-Link, WD HDD(Weldstern Digital), AMD PROC),процессором.. Взаимодействовать с железом помогает ОС. 
ОС - большая прорамма, которая  с помощью микропрограмм(драйверов) взаимодействует с различными частями железа.  ТО ОС - это набор обстракций для удобного программирования. Пользователь взаимодействует с железом по средствам ОС обращаясь к ней через интерфейс пользователя.
Одна из гланых задач ОС -  скрыть от пользователя аппаратнное обеспечение и драйвера и взамен открыть удобные абстракции,  интерфейс.
Самые поппулярные ОС -  Android(для приложений), Windows(среди десктопных ОС), Linux(для серверных решений)

ОС Windows - работает с железом при помощи микропрограмм(драйверов=микро ядра) используюя уровень абстракции от железа(Harware abstraction layer=HAL).  
Состоит из:
1) гибридное ядро ()
2) драйверов
3) разных сервисов- упраляющих программ для компонентов ОС: 

 3.1. I/O mnger(input/output managers)  - чтение и запись данных; 3.2.  power mnger - питание, 3.3. VMM(Virtual mamory mnger) - менеджер виртуальной памяти, 3.4. Process mnger- менеджер процессов, и тд
